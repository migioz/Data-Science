```python
dataset_link = 'AmesHousing.csv'
df = pd.read_csv(dataset_link)

sns.set(style="whitegrid")

df['Total Bathrooms'] = df['Full Bath'] + 0.5 * df['Half Bath']

df['Total Porch Area'] = df['Open Porch SF'] + df['Enclosed Porch'] + df['3Ssn Porch'] + df['Screen Porch']

df['Outdoor Area'] = df['Wood Deck SF'] + df['Total Porch Area']

df['Age at Sale'] = df['Yr Sold'] - df['Year Built']

color_palette = sns.cubehelix_palette(rot=-.4, as_cmap=True)

font_bold = {'weight': 'bold'}

label_color_tosca = {'color': '#008080'}

plt.figure(figsize=(12, 6))
sns.histplot(df, x='SalePrice', bins=30, kde=True, multiple='stack', palette='viridis')
plt.title('Distribution of Sale Prices by House Style', fontdict={**font_bold, **label_color_tosca})
plt.xlabel('Sale Price', fontdict={**font_bold, **label_color_tosca})
plt.ylabel('Frequency', fontdict={**font_bold, **label_color_tosca})
plt.legend(title='House Style', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.legend(['Frequency', 'Sale Price'], loc='upper right')  # Adjust the legend labels accordingly
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df['SalePrice'], bins=30, kde=True, color='darkblue')
plt.title('Distribution of Sale Prices', fontdict={**font_bold, **label_color_tosca})
plt.xlabel('Sale Price', fontdict={**font_bold, **label_color_tosca})
plt.ylabel('Frequency', fontdict={**font_bold, **label_color_tosca})
plt.legend(title='Overall Quality', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='House Style', y='SalePrice', data=df, palette='Set3')
plt.title('House Style vs. Sale Price', fontdict={**font_bold, **label_color_tosca})
plt.xlabel('House Style', fontdict={**font_bold, **label_color_tosca})
plt.ylabel('Sale Price', fontdict={**font_bold, **label_color_tosca})
plt.xticks(rotation=45)
plt.suptitle('House Style vs. Sale Price with Fireplace Quality', y=1.05, fontdict={**font_bold, **label_color_tosca})
plt.show()

plt.figure(figsize=(12, 6))
sns.regplot(x='Gr Liv Area', y='SalePrice', data=df, scatter_kws={'s': df['Overall Qual']*5}, line_kws={'color': 'red'})
plt.title('Outdoor Area vs. Sale Price with Regression Line and Overall Quality Size', **font_bold, **label_color_tosca)
plt.xlabel('Outdoor Area', **font_bold, **label_color_tosca)
plt.ylabel('Sale Price', **font_bold, **label_color_tosca)
plt.legend(['Overall Quality'])

plt.show()
```
